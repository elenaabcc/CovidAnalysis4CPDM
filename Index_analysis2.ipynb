```python
import pandas as pd
import numpy as np
import plotly.graph_objs as go
import plotly.io as pio
import plotly.express as px
pio.renderers.default='plotly_mimetype+notebook'
from ipywidgets import widgets as wg
import yfinance as yf
import requests
```

```python

r = requests.get('https://finance.yahoo.com/world-indices/')
c = r.content
df_list = pd.read_html(c)

majorStockIdx = df_list[0]
majorStockIdx.head()
```

```python
# ^GSPC is the symbol of S&P 500 in Yahoo! Finance
tickerData = yf.Ticker('^GSPC')
```

```python
#We Set the period we want to analyze
tickerDf1 = tickerData.history(period='1d', start='2019-1-1', end='2022-1-1')
```

```python
# Now it’s time to collect historical data of ALL stock indices.
stock_list = []
for s in majorStockIdx.Symbol: # iterate for every stock indices
    # Retrieve data from Yahoo! Finance
    tickerData = yf.Ticker(s)
    tickerDf1 = tickerData.history(period='1d', start='2019-1-1', end='2022-1-1')
    # Save historical data 
    tickerDf1['ticker'] = s # don't forget to specify the index
    stock_list.append(tickerDf1)
# Concatenate all data
msi = pd.concat(stock_list, axis = 0)
msi
```
