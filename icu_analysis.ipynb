```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.io as pio
import plotly.express as px
pio.renderers.default='plotly_mimetype+notebook'
```

```python
icu_data=pd.read_csv(r'C:\Users\strag\Documents\GitHub\CovidAnalysis4CPDM\PROGETTO\icu dataframe.csv')
icu_data
```

```python
#MEAN NUMBER OF COVID PATIENTS IN HOSPITAL AND ICU OCCUPANCY FROM THE BEGINNING TO NOW
icu_italy=icu_data[icu_data['country']=='Italy']
my_cond=['Daily ICU occupancy', 'Daily hospital occupancy']
icu_italy=icu_italy[icu_italy['indicator'].isin(my_cond)]
icu_italy[['Year', 'Month', 'Week']]=icu_italy['date'].str.split('-', expand=True)
icu_italy.head()
```

```python
#ICU AND HOSPITAL OCCUPANCY IN 2021
icu_2021=icu_italy[icu_italy.Year=='2021']
table_2021=icu_2021.pivot_table(index='Month', columns='indicator', values='value', aggfunc='mean')
table_2021.plot(kind='bar', title='ICU and hospital occupancy in 2021', color=['red', 'yellow'])
plt.ylabel('Mean')
plt.savefig('ICU and H.O. in 2021.jpeg', dpi=300, bbox_inches='tight')
```

```python
icu_2020=icu_italy[icu_italy.Year=='2020']
table_update=icu_2020.pivot_table(index='Month', columns='indicator', values='value', aggfunc='mean')
table_update.plot(kind='bar', title='ICU and hospital occupancy in 2020', color=['red', 'yellow'])
plt.ylabel('Mean')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.savefig('ICU and H.O. in 2020.jpeg',dpi=300, bbox_inches='tight')
```
