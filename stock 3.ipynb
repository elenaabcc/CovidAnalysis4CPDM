```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
#!pip install yfinance
import yfinance as yf
%matplotlib inline
```

```python
start = "2019-01-01"
end = '2022-01-01'
ftsemib = yf.download('FTSEMIB.MI',start,end)
sp500 = yf.download('^GSPC',start,end)
nasdaq = yf.download('^IXIC',start,end)
dow30 = yf.download('^DJI',start,end)
```

```python
#exploratory analysis
ftsemib['Open'].plot(label = 'FTSEMIB.MI', figsize = (15,7))
sp500['Open'].plot(label = "^GSPC")
nasdaq['Open'].plot(label = '^IXIC')
dow30['Open'].plot(label = '^DJI')
plt.title('Indeex Prices of FTSEMIB, S&P 500, Nasdaq, Dowjones 30')
```

```python
ftsemib['Volume'].plot(label = 'FTSEMIB.MI', figsize = (15,7))
sp500['Volume'].plot(label = "^GSPC")
nasdaq['Open'].plot(label = '^IXIC')
dow30['Volume'].plot(label = '^DJI')
plt.title('Volume of Stock traded')
plt.legend()
```

```python
ftsemib['MarktCap'] = ftsemib['Open'] * ftsemib['Volume']
sp500['MarktCap'] = sp500['Open'] * sp500['Volume']
nasdaq['MarktCap'] = nasdaq['Open'] * nasdaq['Volume']
dow30['MarktCap'] = dow30['Open'] * dow30['Volume']
ftsemib['MarktCap'].plot(label = 'FTSEMIB.MI', figsize = (15,7))
sp500['MarktCap'].plot(label = "^GSPC")
nasdaq['Open'].plot(label = '^IXIC')
dow30['MarktCap'].plot(label = '^DJI')
plt.title('Markt Cap')
plt.legend()
```
