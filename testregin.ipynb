```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
test=pd.read_csv("./owid-covid-data.csv")
test.columns
```

```python
test=test.query("location == 'Italy'")
test
```

```python
fig=go.Figure(go.Scatter(x=test['date'], y=test['total_vaccinations']))
fig


```

```python
fig.add_trace(
    go.Bar(x=test['date'], y=(test['total_vaccinations']),
    marker=dict(color='green', coloraxis="coloraxis")
    ),
    row=1, col=1
)

fig.add_trace(
    go.Bar(x=test['date'], y=(test['new_deaths'])),
    row=2, col=1)
```
